```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
url = r'https://www.abs.gov.au/statistics/labour/employment-and-unemployment/labour-force-australia-detailed/nov-2021/EQ12.xlsx'
df = pd.read_excel(url,sheet_name="Data 1")
```

```python
df.shape
```

```python
df.columns
```

```python
df_new=df.iloc[2:,:]
df_new.head()
```

```python
new_header=df_new.iloc[0]
df_new=df_new.iloc[1:,:]
df_new.columns=new_header
df_new.head()
```

```python
df_new=df_new.iloc[:,:7]
```

```python
df_new=df_new.reset_index()
```

```python
df_new['date'] = pd.to_datetime(df_new['Mid-quarter month'], format='%Y-%m-%d')
```

```python
df_new['date'].unique()
```

```python
df_new1=df_new[(df_new["date"]=="2021-11-01") | (df_new["date"]=="2020-11-01") | (df_new["date"]=="2019-11-01") | 
              (df_new["date"]=="2018-11-01") | (df_new["date"]=="2017-11-01") | (df_new["date"]=="2016-11-01")]
```

```python
df_new1
```

```python
df_new1=df_new1[df_new1["Industry division of main job: ANZSIC (2006) Rev.2.0"].isin({"Manufacturing","Information Media and Telecommunications","Professional, Scientific and Technical Services"})]
```

```python
df_new1_tech=df_new1[df_new1["Industry division of main job: ANZSIC (2006) Rev.2.0"]=="Professional, Scientific and Technical Services"]
df_new1_tech_15_24=df_new1_tech[(df_new1_tech["Age"]=="15-19 years") | (df_new1_tech["Age"]=="20-24 years")]
df_new1_tech_25_34=df_new1_tech[(df_new1_tech["Age"]=="25-29 years") | (df_new1_tech["Age"]=="30-34 years")]
df_new1_tech_35_44=df_new1_tech[(df_new1_tech["Age"]=="35-39 years") | (df_new1_tech["Age"]=="40-44 years")]
df_new1_tech_45_54=df_new1_tech[(df_new1_tech["Age"]=="45-49 years") | (df_new1_tech["Age"]=="50-54 years")]
df_new1_tech_55_64=df_new1_tech[(df_new1_tech["Age"]=="55-59 years") | (df_new1_tech["Age"]=="60-64 years")]
df_new1_tech_65_over=df_new1_tech[(df_new1_tech["Age"]=="65 years and over")]
```

```python
df_new1_tech
```

```python
x=np.linspace(-20, 20, 1000)
plt.plot( 'Mid-quarter month', "Employed full-time ('000)", data=df_new1_tech_15_24, linestyle="-", markerfacecolor='blue', markersize=12, color='skyblue', linewidth=1,label="15-24 years")
plt.plot( 'Mid-quarter month', "Employed full-time ('000)", data=df_new1_tech_25_34, linestyle="--", markerfacecolor='blue', markersize=12, color='red', linewidth=1,label="25-34 years")
plt.plot( 'Mid-quarter month', "Employed full-time ('000)", data=df_new1_tech_35_44, linestyle="-.", markerfacecolor='blue', markersize=12, color='black', linewidth=1,label="35-44 years")
plt.plot( 'Mid-quarter month', "Employed full-time ('000)", data=df_new1_tech_45_54, linestyle=":", markerfacecolor='blue', markersize=12, color='green', linewidth=1,label="45-54 years")
plt.plot( 'Mid-quarter month', "Employed full-time ('000)", data=df_new1_tech_55_64, linestyle=":", markerfacecolor='blue', markersize=12, color='olive', linewidth=1,label="55-64 years")
plt.plot( 'Mid-quarter month', "Employed full-time ('000)", data=df_new1_tech_65_over, linestyle=":", markerfacecolor='blue', markersize=12, color='blue', linewidth=1,label="65 and over years")
plt.legend(bbox_to_anchor=(1, 0.8))
plt.show()
```
